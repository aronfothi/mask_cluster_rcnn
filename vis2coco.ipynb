```python
import json
import os
```

```python
vis_ann_path = '/media/hdd/aron/rats/human/8597db4/training_data/annotations/instances_val.json'
coco_ann_path = '/media/hdd/aron/rats/human/8597db4/training_data/annotations/coco_instances_val.json'
```

```python
with open(vis_ann_path) as f:
    vis_ann = json.load(f)
```

```python
vis_ann.keys()
```

```python
coco_ann = dict(info=vis_ann['info'], licenses=vis_ann['licenses'], categories=vis_ann['categories'])
images = []
annotations = []
```

```python
coco_ann['categories']
```

```python
image_id = 0
annotation_id = 0
for vid in vis_ann['videos']:
    vid_id = vid['id']
    #print(vid['file_names'])
    #print(vid_id)
    for ann in vis_ann['annotations']:
        if ann['video_id'] == vid_id:            
            for i, (seg, bbox, area) in enumerate(zip(ann['segmentations'], ann['bboxes'], ann['areas'])):
                annotation = dict(id=annotation_id,
                                   image_id=image_id+i,
                                   segmentation=seg,
                                   area=area,
                                   bbox=bbox,
                                   category_id=1,
                                   iscrowd=0)
                annotations.append(annotation)
                annotation_id += 1
        
    for file_name in vid['file_names']:
        file_name = os.path.relpath(file_name, '/media/hdd/aron/rats/human/8597db4/training_data/images/')
        image = dict(id=image_id,
                     width=640,
                     height=420,
                     file_name=file_name,
                     license=1,
                     flickr_url='http://nipg13.inf.elte.hu:8000/rats/images/' + file_name,
                     coco_url='http://nipg13.inf.elte.hu:8000/rats/images/' + file_name,
                     date_captured='')
        images.append(image)
        image_id += 1
        
```

```python
coco_ann['images'] = images
coco_ann['annotations'] = annotations
```

```python
with open(coco_ann_path, 'w') as json_file:
    json.dump(coco_ann, json_file)
```
